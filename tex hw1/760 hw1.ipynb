```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mean = np.array([0,0])
cov = np.array([[1,0],[0,1]])
s = np.random.multivariate_normal(mean, cov, 100)
x = []
y = []
for i in range(len(s)):
    x.append(s[i][0])
    y.append(s[i][1])
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.scatter(x,y)
plt.savefig("1")
```
